# Do Justices have a bias in what party they retire under?

```python
Democratdied=17
Republicandied=20
Democraticrepublicandied=5
Whigdied=2
Federalistdied=1
Nationaluniondied=1

Democratretired=18
Republicanretired=32
Georgewashingtonretired=4
```

```python
labels = 'Democratic President', 'Republican President'
died = [Democratdied, Republicandied]
retired= [Democratretired, Republicanretired]
colors = ['b','r']

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d})".format(pct, absolute)

fig = plt.figure(1, figsize=(3,6))
fig.patch.set_facecolor('white')
ax1=fig.add_subplot(2,1,1)
ax1.set_title("What fraction of SCOTUS Justices die under each party?",fontsize=11)
ax1.pie(died,labels=labels,colors=colors, autopct=lambda pct: func(pct, (18,20)),shadow=True, startangle=95)
ax1.axis('equal')

ax2=fig.add_subplot(2,1,2)
ax2.set_title("What fraction of SCOTUS Justices resign under each party?",fontsize=11)
ax2.pie(retired,labels=labels,colors=colors,autopct=lambda pct: func(pct, (18,33)),shadow=True, startangle=110,textprops=dict(color="k"))
ax2.axis('equal')
plt.show()
```
